```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Tsunami Data.csv")
```

```python
missing_percent = df.isnull().mean() * 100
missing_percent
```

```python
threshold = 0.90
df = df.loc[:, df.isnull().mean() < threshold]
df
```

```python
df.info()
```

```python
df['Mo'].fillna(df['Mo'].median(), inplace=True)
df['Dy'].fillna(df['Dy'].median(), inplace=True)
df['Hr'].fillna(0, inplace=True)
df['Mn'].fillna(0, inplace=True)
df['Sec'].fillna(0, inplace=True)
df['Latitude'].fillna(df['Latitude'].mean(), inplace=True)
df['Longitude'].fillna(df['Longitude'].mean(), inplace=True)
df['EarthquakeNanMagnitude'].fillna(df['EarthquakeNanMagnitude'].median(), inplace=True)
df['MaximumNanWaterNanHeightNan(m)'].fillna(df['MaximumNanWaterNanHeightNan(m)'].median(), inplace=True)
df['TsunamiNanMagnitudeNan(Iida)'].fillna(df['TsunamiNanMagnitudeNan(Iida)'].median(), inplace=True)
df['TsunamiNanIntensity'].fillna(df['TsunamiNanIntensity'].median(), inplace=True)
df['Deaths'].fillna(0, inplace=True)
df['TotalNanDeaths'].fillna(0, inplace=True)
df['TotalNanInjuries'].fillna(0, inplace=True)
df['TotalNanDamageNan($Mil)'].fillna(0, inplace=True)
df['TotalNanHousesNanDestroyed'].fillna(0, inplace=True)
```

```python
df.info()
```

```python
df['DeathNanDescription'].dtype
df['DeathNanDescription'].unique()
```

```python
cols = [
    'DeathNanDescription', 'DamageNanDescription',
    'TotalNanDeathNanDescription', 'TotalNanInjuriesNanDescription',
    'TotalNanDamageNanDescription', 'TotalNanHousesNanDestroyedNanDescription',
    'TotalNanHousesNanDamagedNanDescription'
]

for col in cols:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col].fillna(0, inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# If your df might be a slice, create a clean copy first
df = df.copy()

# Then run the datetime creation
df_rename = df.rename(columns={
    'Year': 'year',
    'Mo': 'month',
    'Dy': 'day',
    'Hr': 'hour',
    'Mn': 'minute',
    'Sec': 'second'
})

df.loc[:, 'Datetime'] = pd.to_datetime(df_rename[['year', 'month', 'day', 'hour', 'minute', 'second']], errors='coerce')

print(df[['Year', 'Mo', 'Dy', 'Hr', 'Mn', 'Sec', 'Datetime']].head())
```

```python
df.rename(columns={
    'Mo': 'Month',
    'Dy': 'Day',
    'Hr': 'Hour',
    'Mn': 'Minute',
    'Sec': 'Second'
}, inplace=True)
```

```python
df.info()
```

```python
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Day'] = df['Datetime'].dt.day
df['Hour'] = df['Datetime'].dt.hour
df['Weekday'] = df['Datetime'].dt.dayofweek  # Monday=0, Sunday=6
df['IsWeekend'] = df['Weekday'].isin([5, 6]).astype(int)
```
